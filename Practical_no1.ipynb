```python
from ast import operator
print("ARITHMATIC OPERATORS")
a=int(input("value of a is"))
b=int(input("value of b is"))
add=a+b
print(f"addition of {a} and {b} is", add)
subtract=a-b
print(f"subtraction of {a} and {b} is", subtract)
multiplication=a*b
print(f"Multiplication of {a} and {b} is",multiplication)
division=a/b
print(f"Division of {a} and {b} is", division)
modulous=a%b
print(f"Modulous of {a} and {b} is", modulous)
power=a^b
print(f"power of {a} and {b} is", power)
floatdiv=a//b
print(f"Float division of {a} and {b} is", floatdiv)
```

```python
print("RELATIONAL OPERATORS")
a=int(input("value of a"))
b=int(input("value of b"))
print(f"{a} is less than {b}?:",a<b)
print(f"{a} is greater than {b}?:",a>b)
print(f"{a} less than equal to {b}?:",a<=b)
print(f"{a} is greater than equal to {b}?:",a>=b)
print(f"{a} is not equal to {b}?:",a!=b)
print(f"{a} is equal to {b}?:",a==b)
```

```python
print("LOGICAL OPERATORS")
a=True
b=False
print(f"{a} and {b}?:",a and b)
print(f"{a} or {b}?:", a or b)
print(f"not {b}?:", not b)
```

```python
print("Bitwise Operator-")
a = int(input("Enter the value of a: "))
b = int(input("Enter the value of b: "))
print(f"The value of {a} & {b} is = {a & b}")
print(f"The value of {a} | {b} is = {a | b}")
print(f"The value of {a}^{b} is = {a^ b}")
print(f"The value of ~{a} is = {~a}")
print(f"The value of {a} <<< {b} is = {a << b}")
print(f"The value of {a} >> {b} is = {a >> b}")
```
