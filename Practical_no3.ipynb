```python
a=int(input("Enter a no."))
b=int(input("Enter a no."))
choose=input("Enter the operation no.")
if(choose == '+'):
  a+=b
elif(choose == '-'):
  a-=b
  print(f"{a} is addition of a and b")
  print(f"{a} is substraction of a and b")
elif(choose == '*'):
  a*=b
  print(f"{a} is product of a and b")
elif(choose == '/'):
  a/=b
  print(f"{a} is division of a and b")
elif(choose == '%'):
  a%=b
  print(f"{a} is modulus of a and b")
else:
  print("Invalid Operation")
```
